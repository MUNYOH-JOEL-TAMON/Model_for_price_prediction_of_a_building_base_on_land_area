```python
import pandas as pd 
import numpy as np 
```

```python
data = {
    'AREA': 
    [
        1000, 2600, 2900, 3000, 3200, 3600, 4000, 5500, 6300, 8700, 9300, 10200
    ], 
    'PRICES': 
    [
        '3M', '5.5M', '5.59M', '5.65M', '6.1M', '6.8M', '7.75M', '8.12M', '9.27M', '11.47M', '13.22M', '15M'
    ]
}

df = pd.DataFrame(data)
df
```

```python
df['PRICES'] = df['PRICES'].str.replace('M', '').astype(float)*1_000_000 
df
```

```python
import matplotlib.pyplot as plt 

plt.figure(figsize=(8,5))

plt.scatter(df['AREA'], df['PRICES'], color='green')
plt.title('Area in (sq ft) Vs Price (in millions)')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price (in millions)')

plt.ticklabel_format(style='plain', axis='y')
plt.grid(True)

plt.show()
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split

X = df[['AREA']]
y = df['PRICES']

model = LinearRegression()
model.fit(X,y)

print('Model Training Completed')
```

```python
new_houses = pd.DataFrame({
    'AREA': 
    [
        1200, 9000, 1100
    ]
})

predicted_price = model.predict(new_houses)
for area, price in zip(new_houses['AREA'], predicted_price):
    print(f"A building with {area} sq ft is predicted to cost: {price:,.2f} FCFA")
```
